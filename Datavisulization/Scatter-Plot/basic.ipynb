# Scatter plot

### 1. Bivariate Analysis

### To find corelation between two quantities

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(-10,10,50)
y = 10 * x + 3 + np.random.randint(0,300,50)
```

### .scatter() function

```python
plt.scatter(x,y)
```
