# Working with "sharma-kohli.csv" data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
batsman = pd.read_csv('/Users/mohit/Documents/matplotlib/datasets/sharma-kohli.csv')
print(batsman)
```

```python
# plotting graph of a particular batsman ( Virat kohli)

# single plotting
plt.plot(batsman['index'],batsman['V Kohli'])
```

## Multiple plotting (without lable)

```python
# multiple plotting 
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

## Lable title (plotting with labels)

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### color (hex) 

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')

```

### line(width and style) --dashed  and marker(size)

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dashed')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dotted')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```
