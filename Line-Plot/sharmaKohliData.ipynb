# Working with "sharma-kohli.csv" data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
batsman = pd.read_csv('datasets/sharma-kohli.csv')
print(batsman)
```

```python
# plotting graph of a particular batsman ( Virat kohli)

# single plotting
plt.plot(batsman['index'],batsman['V Kohli'])
```

## Multiple plotting (without lable)

```python
# multiple plotting 
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

## Lable title (plotting with labels)

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### color (hex) 

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')

```

### line(width and style) -- linestyle = 'dashed'

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dashed')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### line(width and style) ....  linestyle = 'dotted'

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dotted')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### line(width and style) -.-.-.  linestyle = 'dashdot'

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dashdot')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### linewidth

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='solid',linewidth=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='solid',linewidth=3)


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### marker = 'D' (D = diamond)

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dashdot',marker='D',markersize=8)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dashdot',marker='D',markersize=6)


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### legend function. ( legend ->location )

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dashdot',marker='D',markersize=8,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dashdot',marker='o',markersize=6,label='Rohit')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# legend function
plt.legend()
```

### limiting axes

```python
# assume data

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
price = [30000, 20000, 45000, 42000, 50000, 59000, 70000, 60000, 79000, 90000, 8000000]


plt.plot(years,price)

# defining range
plt.ylim(0,75000)
# we can set limit to x axis too using .xlim()
```

### .grid(). function

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dashdot',marker='D',markersize=8,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dashdot',marker='o',markersize=6,label='Rohit')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# legend function
plt.legend()
# gird
plt.grid()
```

### .show() function ---- this the the main function which is responsible for plotting graph

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle='dashdot',marker='D',markersize=8,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle='dashdot',marker='o',markersize=6,label='Rohit')


#graph title
plt.title('Virat Kohli VS Rohit Sharma Runs Comparison')

# lables
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# legend function
plt.legend()

# .grid()
plt.grid()

# .show()
plt.show()
```
