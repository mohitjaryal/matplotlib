## Working with 'gayle-175.csv' data set

### Pie chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gl = pd.read_csv('/datasets/gayle-175.csv')
```

```python
# pie chart

plt.pie(gl['batsman_runs'],labels=gl['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# colors
plt.pie(gl['batsman_runs'],labels=gl['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','red','purple'])
plt.show()
```

### explode & shadow

```python
plt.pie(gl['batsman_runs'],labels=gl['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','red','purple'],
explode=[0.1,0,0,0,0,0.1],shadow=True)
plt.show()
```
