## 2D line plot

### Used for 
### Bivariate Analysis
### categorical -> numerical and numerical -> numerical
### Use case - Time series data

```python
# plotting graph on sharma_kohli dataset

# plot kohli's graph
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'])
```

```python
# plotting graph on sharma_kohli dataset

# plot rohit's graph
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'])
```

```python
# plotting multiple plots on a single graph

plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'])
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'])
```

```python
# lables title
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'])
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'])
plt.title('Rohit Sharma VS Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# colors(hex) and line(width and style) and marker size 
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],color='red',linestyle='solid',linewidth=2)
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],color='blue',linestyle='dotted',linewidth=5)
plt.title('Rohit Sharma VS Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# marker and marker size

plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],color='red',linestyle='solid',linewidth=2,marker='D',markersize=7)
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],color='blue',linestyle='dotted',linewidth=5,marker='o',markersize=10)
plt.title('Rohit Sharma VS Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# legend
# marker and marker size

plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],color='red',linestyle='solid',linewidth=2,marker='D',markersize=5,label='Virat')
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],color='blue',linestyle='dotted',linewidth=5,marker='o',markersize=5,label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
```

```python
# limiting axes

price = [90000, 80000, 70000, 80000, 85000, 4500000]
year  = [2022, 2023, 2024, 2025, 2026, 2027]

plt.plot(year, price, marker='o')

# zoom on normal data
plt.ylim(60000, 100000)

# outlier point 
plt.scatter(2027, 4500000, color='red', label='Outlier (2027)')

plt.legend()
plt.title("Price Trend with Outlier Highlighted")
plt.show()
```

```python
# grids

plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],color='red',linestyle='solid',linewidth=2,marker='D',markersize=5,label='Virat')
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],color='blue',linestyle='dotted',linewidth=5,marker='o',markersize=5,label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
plt.grid()
```

```python
# grids

plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],color='red',linestyle='solid',linewidth=2,marker='D',markersize=5,label='Virat')
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],color='blue',linestyle='dotted',linewidth=5,marker='o',markersize=5,label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
plt.grid()
plt.show()
```

## Bar Chart
### Bivariate Analysis
### Numerical vs Categorical
### Use case - Aggregate analysis of groups

```python
boys = [20,40,23,32,46,41]
colors = ['red','blue','green','yellow','black','orange']


# vertical bar chart
plt.bar(colors,boys,color='black')
plt.xlabel('Colors')
plt.ylabel('Boys')
plt.title('Boys fav color')
plt.show()
```

```python
# horizontal bar chart -> plt.barh()
plt.barh(colors,boys,color='black')
plt.xlabel('Colors')
plt.ylabel('Boys')
plt.title('Boys fav color')
plt.show()
```

```python
# multiple bar chart 

plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
import matplotlib.pyplot as plt

boys = [20,40,23,32,46,41]
colors = [
    'red',
    'blue blue blue blue blue',
    'green green green green green',
    'yellow yellow yellow yellow yellow',
    'black black black black black',
    'orange orange orange orange'
]

plt.bar(colors, boys, color='black')
plt.xlabel('Colors')
plt.ylabel('Boys')
plt.title('Boys fav color')

plt.xticks(rotation=90)   # rotate labels vertically
plt.tight_layout()        # prevents label cutoff

plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'],label='2017')
plt.legend()
plt.show()
```

## Pie Chart
### Univariate/Bivariate Analysis
### We can use Pie Chart for both Categorical and numerical data
### Use case - To find contibution on a standard scale (hr category ka 100% me kitna contribution h)

```python
# working with gayle data set
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# specified colors

plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','red','brown','pink'])
plt.show()
```

```python
# extracting a particular section -> using explode
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0.2])
plt.show() 
```

```python
# shadow
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0.2],shadow=True)
plt.show() 
```
