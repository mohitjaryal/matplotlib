## Performing visulization on "batsman_season_records.csv" dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
best = pd.read_csv('/Users/mohit/Documents/matplotlib/datasets/batsman_season_record.csv')
best
```

### making multiple barchart

```python
# making barchart - 
x = np.arange(best.shape[0])   # positions for players
width = 0.25                   # width of each bar

plt.bar(x - width, best['2015'], width=width, label='2015')
plt.bar(x,         best['2016'], width=width, label='2016')
plt.bar(x + width, best['2017'], width=width, label='2017')

plt.title('Top Batsmen Performance (2015â€“2017)')
plt.xlabel('Batsmen')
plt.ylabel('Runs')
plt.legend()
plt.tight_layout()
plt.show()
```
