## Performing visulization on "batsman_season_records.csv" dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
best = pd.read_csv('datasets/batsman_season_record.csv')
best
```

### making multiple barchart

```python
# making barchart - 
x = np.arange(best.shape[0])   # positions for players
width = 0.25                   # width of each bar

plt.bar(x - width, best['2015'], width=width, label='2015',color='red')
plt.bar(x,         best['2016'], width=width, label='2016',color='pink')
plt.bar(x + width, best['2017'], width=width, label='2017',color='black')

plt.title('Top Batsmen Performance (2015â€“2017)')
plt.xticks(x,best['batsman'])
plt.xlabel('Batsmen')
plt.ylabel('Runs')
plt.legend()
plt.tight_layout()
plt.show()
```

### A problem occurs if the name of any column is very long then we can use .xticks(rotation)

```python
childrens = [10,20,35,60,40,90]
colors = ['reddddddddddddd','greennnnnnnnnnnn','yellowwwwwwww','brownnnnnnnn','blackkkkkkkkkkkkkk','pinkkkkkkkkk']


plt.bar(colors,childrens,color='black')
plt.xticks(rotation='vertical')
```

### Stacked barchart

```python
# stacked barchart
plt.bar(best['batsman'], best['2015'],color='pink', label='2015')
plt.bar(best['batsman'], best['2016'],color='red', bottom=best['2015'], label='2016')
plt.bar(best['batsman'], best['2017'],color='black', bottom=best['2015'] + best['2016'], label='2017')

plt.title('Top Batsmen Performance (2015â€“2017)')
plt.xlabel('Batsmen')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
