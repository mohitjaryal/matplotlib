## Basic demonstration of BarChart 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# basic barchart
childrens = [10,20,35,60,40,90]
colors = ['red','green','yellow','brown','black','pink']
```

```python
# creating barchart - (vertical barchart)
plt.bar(colors,childrens,color='black')
plt.title('Color preference by childrens')
plt.xlabel('Colors')
plt.ylabel('Childrens')
```

```python
# creating barchart - (horizontal barchart) using barh()
plt.barh(colors,childrens,color='black')
plt.title('Color preference by childrens')
plt.xlabel('Colors')
plt.ylabel('Childrens')
```
