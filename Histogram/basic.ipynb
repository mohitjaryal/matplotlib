## Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# basic histogram
scores = [32,45,56,10,15,27,61]

# Create histogram
plt.hist(scores, bins=[10,25,40,55,70], color='black')

plt.title("Distribution of Student Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

plt.show()
```
