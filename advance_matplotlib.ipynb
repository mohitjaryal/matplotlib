# Advance Matplotlib

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing iris data set
iris = pd.read_csv('data/iris.csv')

# batter data set
batter = pd.read_csv('data/batter.csv')

#importing ipl ball by ball data set
delivery = pd.read_csv('data/IPL_Ball_by_Ball_2008_2022.csv')
```

### Colored Scatter Plot

```python
# reasigining iris color 
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1 ,'Iris-virginica':2})
iris['Species']
```

```python
# findind relationship b/w SepalLengthCm and PetalLengthCm	
plt.scatter(iris['PetalLengthCm'],iris['SepalLengthCm'],c=iris['Species']) # alpha -> another parameter used to adjust opacity e.g alpha-0.1 and c means color
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.title('SepalLength and PetalLength')
plt.colorbar()
plt.show()
```

### Plot Size

```python
# increasing size of graph

# figure
plt.figure(figsize=(6,4)) # (width, height)

plt.scatter(iris['PetalLengthCm'],iris['SepalLengthCm'],c=iris['Species']) 
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.title('SepalLength and PetalLength')
plt.colorbar()
plt.show()
```

## Annotaions

```python
# working with batter data set
sample_batter = batter.head(100).sample(25,random_state=5)
sample_batter
```

```python
# plotting
plt.figure(figsize=(12,6))
plt.scatter(sample_batter['avg'],sample_batter['strike_rate'])

# using for loop
for i in range(sample_batter.shape[0]):
    plt.text(sample_batter['avg'].values[i],sample_batter['strike_rate'].values[i],sample_batter['batter'].values[i])
```

## Horizontal and Vertical lines 
### (plotting horizontal and vertical lines on a graph)

```python
plt.figure(figsize=(12,6))
plt.scatter(sample_batter['avg'],sample_batter['strike_rate'])

# horizontal line
plt.axhline(130,color='red')

# vertical line
plt.axvline(30,color='blue')
# using for loop
for i in range(sample_batter.shape[0]):
    plt.text(sample_batter['avg'].values[i],sample_batter['strike_rate'].values[i],sample_batter['batter'].values[i])
```

## Sub plots

```python
# another way to plot graph

fig, ax = plt.subplots(figsize=(9,6))

ax.scatter(batter['avg'],batter['strike_rate'])
# setting title
ax.set_title('Avg and Strike Rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
# plotting avg vs sr and runs vs avg
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,4))

# first axis
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title('Avg VS Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg VS Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# plotting avg vs sr and runs vs avg
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
```

```python
# another way to plot sub plots by creating figure object
fig = plt.figure() # figure object

ax1 = fig.add_subplot(2,1,1) # 2,2,1 means -> 2 rows, 1 column and 1st graph
ax1.scatter(batter['avg'],batter['strike_rate'],color='red') 

ax2 = fig.add_subplot(2,1,2) # 2,1,2 means -> 2 rows, 2 columns and 2nd graph
ax2.hist(batter['runs'])
```

## 3D Plots using Matplotlib

### 3D Scatter plot

```python
# figure object
fig = plt.figure(figsize=(9,5))

# axis object
ax = plt.subplot(projection='3d')

# scatter3d funcition
ax.scatter3D(batter['runs'],batter['avg'],batter['strike_rate']) # in the sctter3d function we'll provide x,y and z axis

# title
ax.set_title('IPL batsman analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
```

### 3D Line plot

```python
# creating a small data
x =[0,1,5]
y=[0,10,13]
z=[0,13,20]


# 3d line plotting
# figure object
fig = plt.figure(figsize=(9,5))

# axis object
ax = plt.subplot(projection='3d')

# scatter3d funcition
ax.scatter3D(x,y,z)
ax.plot3D(x,y,z,color='red')
```

### 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

# forming a gird using meshgrid
xx, yy = np.meshgrid(x,y)

# calculating z
z = xx**2 + yy**2

# plotting
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
g = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(g)
```

### Contour Plot
##### It represents a 3D graph in 2D

```python
# contour plot -> ( suppose it like you are viewing the graph from the top)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
g = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(g)
```

```python
# contourf plot
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
g = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(g)
```

## Heatmap

```python
# working with delivery data set

# filtering
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]

# piviot table
grid = temp_df.pivot_table(index='overs',columns=['ballnumber'],values='batsman_run',aggfunc='count')

# plotting heatmap
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.colorbar()
```
